### Dhaka Stock Exchange dataset from 2000 to 2025

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/drive/MyDrive/Personal notebooks/Datasets/dhaka_stock_exchange/merged_data.csv')
df.head()
```

```python
df.tail()
```

```python
df.shape
```

How many unique companies are in the column

```python
df['Scrip'].unique()
```

```python
companies=df['Scrip'].nunique()
companies
```

```python
df_Grameen=df[df['Scrip']=='GP']
df_Grameen.head()
```

```python
df_Grameen['Date']=pd.to_datetime(df_Grameen['Date'])
```

```python
avg_volume_per_year=df_Grameen.groupby(df_Grameen['Date'].dt.year)['Volume'].mean()
plt.figure(figsize=(10,6))
plt.plot(avg_volume_per_year.index,avg_volume_per_year.values,marker='o',linestyle='-')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Average Volume of GP per Year')
plt.xlabel('Year')
plt.ylabel('Average Volume')
plt.show()
```

```python
avg_close_per_year=df_Grameen.groupby(df_Grameen['Date'].dt.year)['Close'].mean()
plt.figure(figsize=(12,8))
plt.plot(avg_close_per_year.index,avg_close_per_year.values,marker='o',linestyle='-')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Average Close Price of GP per Year')
plt.xlabel('Year')
plt.ylabel('Average Close Price')
plt.show()
```

So , from the graph above we can see the closing prices were at it is peak in 2018 . We also can see that the market collapsed in 2011 and we all know more or less about that tragedy

```python
df_Grameen['Day']=df_Grameen['Date'].dt.day
df_Grameen['Month']=df_Grameen['Date'].dt.month
df_Grameen['Year']=df_Grameen['Date'].dt.year
df_Grameen["Day_of_week"]=df_Grameen['Date'].dt.dayofweek
df_Grameen.head()
```

```python
corr=df_Grameen.drop(columns=['Scrip','Date']).corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

### Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test =train_test_split(df_Grameen.drop(columns=['Scrip',"Close",'Date'],axis=1),df_Grameen['Close'],test_size=0.2,random_state=42)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linearmodel=LinearRegression()
```

```python
linearmodel.fit(x_train,y_train)
```

```python
y_pred=linearmodel.predict(x_test)
```

```python
from sklearn.metrics import r2_score , confusion_matrix , classification_report
```

```python
score=r2_score(y_test,y_pred)
print("Accuracy Score :",score)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.xlabel('Actual Close Price')
plt.ylabel('Predicted Close Price')
plt.title(f'Actual vs Predicted Close Price (R² = {score})')
plt.show()
```

```python
mean_absolute_error = np.mean(np.abs(y_test - y_pred))
print("Mean Absolute Error:", mean_absolute_error)
```

```python
y_train_pred=linearmodel.predict(x_train)
train_r2=r2_score(y_train,y_train_pred)
print("Train Accuracy Score :",train_r2)
```

The model is performing unusually well . I suspected that would happen by looking at the correlation matrix . There are redundant features .

### Let us create some new feature

```python
df_Grameen['price_range']=df_Grameen['High']-df_Grameen['Low']
df_Grameen.head()
```

```python
df_Grameen=df_Grameen.drop(columns=['High','Low'])
df_Grameen.head()
```

So , now new correlation matrix:

```python
corr=df_Grameen.drop(columns=['Scrip','Date']).corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.show()
```

```python
x=df_Grameen.drop(columns=['Scrip','Date','Close'])
y=df_Grameen['Close']
```

```python
x_train, x_test, y_train, y_test =train_test_split(x,y,test_size=0.2,random_state=42)
```

Now let us proceed with linear regression again

```python
linearmodel.fit(x_train,y_train)
```

```python
y_pred=linearmodel.predict(x_test)
score=r2_score(y_test,y_pred)
print("Accuracy Score :",score)
```

```python
plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linewidth=2)
plt.xlabel('Actual Close Price')
plt.ylabel('Predicted Close Price')
plt.title(f'Actual vs Predicted Close Price (R² = {score})')
plt.show()
```

```python
mean_absolute_error = np.mean(np.abs(y_test - y_pred))
print("Mean Absolute Error:", mean_absolute_error)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(y_test, y_pred)
print("MAPE:", mape)
```

```python
y_train_pred=linearmodel.predict(x_train)
train_r2=r2_score(y_train,y_train_pred)
print("Train Accuracy Score :",train_r2)
```

Still the model performs exceptionally well . Even though we dropped off some unnecessary features . This would be more clear if I had more unseen data .
